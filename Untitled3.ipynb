```python
#preprocessing phase
import pandas as pd
import seaborn as sns
dataset = pd.read_csv("kc_house_data.csv")
dataset.head()
```

```python
dataset['sqft_living'] == dataset['sqft_above'] + dataset['sqft_basement']
```

```python
dataset.drop(['date','id','zipcode','lat','long','sqft_living'], axis = 1, inplace = True)
dataset
#J'ai supprimÃ© les colonnes dates, id, zipcode, lat et long parce que j'estimais qu'elles n'etaient pas importantes, donc elles ne devaient pas necessairement apparaitre dans ma dat
```

```python
dataset.describe()
```

```python
dataset = dataset[dataset['bedrooms'] < 10]
dataset = dataset[dataset['bathrooms'] < 6.5]
dataset = dataset[dataset['sqft_above'] < 10000]
dataset = dataset[dataset['sqft_basement'] < 10000]
dataset = dataset[dataset['sqft_lot'] < 1.5e6]
dataset = dataset[dataset['floors'] < 3.25]
dataset = dataset[dataset['grade'] < 12]
dataset = dataset[dataset['sqft_living15'] < 6000]
dataset = dataset[dataset['sqft_lot15'] < 600000]
dataset
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics

y = dataset['price'].values
x = dataset.drop('price', axis=1).values
x_train, x_test, y_train,  y_test = train_test_split(x, y, test_size=0.20, random_state=30)
model=LinearRegression()   #build linear regression model
model.fit(x_train,y_train)  #fitting the training data
predicted=model.predict(x_test) #testing our modelâ€™s performance


print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```
